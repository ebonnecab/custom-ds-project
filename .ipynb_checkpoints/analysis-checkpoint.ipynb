## Cleaning Data for Analysis

```python
#drop rows with nans
df = df.dropna()
```

```python
#checking if any columns contain na values
df.isna().any()
```

```python
#change floats to ints
df["Discharges"] = df["Discharges"].astype(int)
df["Indicated CPS Reports"] = df["Indicated CPS Reports"].astype(int)
```

```python
#dropping columns with diff types of care
df.drop(df.columns[[2, 3, 4,5,6,7,8,9,10]], axis = 1, inplace = True) 
```

## Descriptive Statistics of Indicated CPS Reports for NYS

What is the average number of indicated CPS reports in NYS?

```python
cps_reports = np.array(df["Indicated CPS Reports"])
def get_mean(dataset):
    return sum(dataset)/len(dataset)

get_mean(cps_reports)
```

What is the median number of indicated CPS reports in NYS?

```python
def get_median(dataset):
    count = len(dataset)
    
    if count < 1:
        return None
    if count % 2 == 1:
        return dataset[(count-1)/2]
    else:
        return (dataset[(count-1)/2] + dataset[(count+1)/2] ) / 2
    
get_median(cps_reports)
```

Grouping the dataframe by county and year to answer some questions!

```python
group = df.groupby(['County'])
group.first()
```

What is the total, mean, and standard deviation of indicated CPS reports per county?

```python
descript = group['Indicated CPS Reports'].agg([np.sum, np.mean, np.std])
descript
```

What is the total, mean, standard deviation by year?

```python
group2 = df.groupby(['Year'])
descript2 = group2['Indicated CPS Reports'].agg([np.sum, np.mean, np.std])
descript2
```

```python
sns.regplot(x="Year", y="Indicated CPS Reports", fit_reg=False, data=df)
```

With the plot above I am able to see that most counties stay within a certain range but there are a few with a much higher of number reports than others. Those counties should be investigated further. This could be due to a bigger population or how many children are in their care overall.

```python
#This plot is very similar to the one above but it accounts for the county as the hue
sns.lmplot(x="Year", y="Indicated CPS Reports", fit_reg=False, data=df, hue="County")
```

Now we are able to see that it is actually one county with specifically high CPS report rates and I believe it is New York City.

```python
#Visualized data without NYC to show how its high rates were skewing the data
no_nyc = df[df.County != 'NEW YORK CITY']
sns.regplot(x="Year", y="Indicated CPS Reports", fit_reg=False, data=no_nyc)
```
